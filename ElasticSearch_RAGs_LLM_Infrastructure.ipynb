```python
index_name = "search-danituntun13"
settings = {
    "settings": {
        "number_of_shards": 1,
        "number_of_replicas": 0
    },
    "mappings": {
        "properties": {
            "text": {"type": "text"},
            "embedding": {"type": "dense_vector", "dims": 384},  # Adjust the dimension with your heart
            "metadata": {"type": "object"}
        }
    }
}

# Create indexes
if not client.indices.exists(index=index_name):
    client.indices.create(index=index_name, body=settings)
    print(f"Index '{index_name}' created")
else:
    print(f"Index '{index_name}' already exists")
```

```python
!pip install PyPDF2
from PyPDF2 import PdfReader

def extract_text_from_pdf(pdf_path):
    reader = PdfReader(pdf_path)
    text = ""
    for page in reader.pages:
        text += page.extract_text()
    return text

pdf_path = "/content/2411.11786v1.pdf"
document_text = extract_text_from_pdf(pdf_path)
print(document_text[:500])
```

```python
def chunk_text(text, chunk_size=200):
    words = text.split()
    for i in range(0, len(words), chunk_size):
        yield " ".join(words[i:i + chunk_size])

chunks = list(chunk_text(document_text))
print(f"Total chunks: {len(chunks)}")
```

```python
!pip install sentence-transformers
from sentence_transformers import SentenceTransformer

# Load the model
model = SentenceTransformer('all-MiniLM-L6-v2')

embedding = model.encode(document_text)
print(embedding)
```

```python
# Add the chunks
for i, chunk in enumerate(chunks):
    doc = {
        "text": chunk,
        "embedding": model.encode(chunk).tolist(),  # Convert the embedding to lists
        "metadata": {"source": "/content/2411.11786v1.pdf", "chunk_id": i}
    }
    client.index(index=index_name, id=i, document=doc)
    print(f"Chunk {i} indexed")
```

```python
query_embedding = model.encode("example search query").tolist()

query = {
    "script_score": {
        "query": {"match_all": {}},
        "script": {
            "source": "cosineSimilarity(params.query_vector, 'embedding') + 1.0",
            "params": {"query_vector": query_embedding}
        }
    }
}

response = client.search(index=index_name, body={"query": query})
print("Search Results:")
for hit in response["hits"]["hits"]:
    print(hit["_source"]["text"])
```
